2. Escreve um programa que lê, um de cada vez, três valores inteiros do teclado e joga cada um em uma variável. Cada valor lido é elevado ao quadrado e dividido por dois. Os valores são, em seguida, somados entre si. Imprima o resultado com uma mensagem "A soma é:" antes de imprimir a soma.

```python
# RESPOSTA:

print("**** Digite somente números inteiros\n")
try:
   valor1 = int(input("Digite o primeiro valor: "))
   valor2 = int(input("Digite o segundo valor: "))
   valor3 = int(input("Digite o terceiro valor: "))
except ValueError:
        print("Número inválido, digite um número inteiro")

def calcQuadrado(valor):
  mult = ((valor ** 2) / 2)
  return mult

soma = int(calcQuadrado(valor1) + calcQuadrado(valor2) + calcQuadrado(valor3))

print("\n")
print("A soma é: ", soma)
```

5. Faça um programa que leia o nome, idade, altura e peso de duas pessoas diferentes e imprima:
- a diferença de altura, peso e idade entre elas
- as médias de: idade, altura e peso entre elas.
Coloque uma mensagem de saída significativa antes das impressões.

```python
# RESPOSTA:

print("\nPrimeira pessoa:")
nome1 = str(input("Digite o nome da primeira pessoa: "))
idade1 = int(input("Digite a idade da primeira pessoa: "))
altura1 = float(input("Digite a altura da primeira pessoa: "))
peso1 = float(input("Digite o peso da primeira pessoa: "))

print("\nSegunda pessoa:")
nome2 = str(input("Digite o nome da segunda pessoa: "))
idade2 = int(input("Digite a idade da segunda pessoa: "))
altura2 = float(input("Digite a altura da segunda pessoa: "))
peso2 = float(input("Digite o peso da segunda pessoa: "))

def calculaDiferenca(altura1, altura2, peso1, peso2, idade1, idade2):
    diferenca_idade = abs(idade1 - idade2)
    diferenca_altura = abs(altura1 - altura2)
    diferenca_peso = abs(peso1 - peso2)

    return diferenca_altura, diferenca_peso, diferenca_idade

def calculaMedia(altura1, altura2, peso1, peso2, idade1, idade2):
    media_idade = ((idade1 + idade2) / 2)
    media_altura = ((altura1+ altura2) / 2)
    media_peso = ((peso1 + peso2) / 2)

    return media_idade, media_altura, media_peso

diferenca_altura, diferenca_peso, diferenca_idade = calculaDiferenca(altura1, altura2, peso1, peso2, idade1, idade2)
media_idade, media_altura, media_peso = calculaMedia(altura1, altura2, peso1, peso2, idade1, idade2)

print("\nDiferença entre a primeira e a segunda pessoa:")
print(f'Diferença de altura: {diferenca_altura} cm')
print(f'Diferença de peso: {diferenca_peso} kg')
print(f'Diferença de idade: {diferenca_idade} anos')

print("\nMédias entre a primeira e a segunda pessoa:")
print(f'Média de idade: {media_idade} anos')
print(f'Média de altura: {media_altura} cm')
print(f'Média de peso: {media_peso} kg')
```
