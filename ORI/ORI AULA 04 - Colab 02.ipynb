3) Escreva um programa que leia um conjunto de 5 fichas, cada uma contendo, a altura e o código do sexo de uma pessoa (código = 1 se for masculino e 2
se for feminino), e calcule e imprima:
- a maior e a menor altura da turma;
- a média de altura das mulheres;
- a média de altura da turma.

```python
conjunto = [
    {"altura": 173, "codsex": 1},
    {"altura": 168, "codsex": 2},
    {"altura": 165, "codsex": 2},
    {"altura": 192, "codsex": 1},
    {"altura": 152, "codsex": 2},
]

# lturas = [d["altura"] for d in conjunto]

alturas = []
alturaMulheres = []

for d in conjunto:
  alturas.append(d["altura"])
  if d["codsex"] == 2:
    alturaMulheres.append(d["altura"])

maior_altura = max(alturas)
menor_altura = min(alturas)

print(f'A maior altura é {maior_altura} cm e a menor é de {menor_altura} cm')

mediaAlturaMulheres = round((sum(alturaMulheres) / len(alturaMulheres)),2)
mediaAlturaGeral = round((sum(alturas) / len(alturas)),2)

print(f'A média de altura das mulheres é de {mediaAlturaMulheres} cm e a média geral é de {mediaAlturaGeral} cm.')
```

4) Fazer um programa que:
- leia um número indeterminado de linhas contendo cada uma a idade de um indivíduo. A última linha, que não entrará nos cálculos, contém o valor da idade igual a zero;
- calcule e escreva a idade média deste grupo de indivíduos.

```python
idades = []

while True:
    idade = int(input("Digite um numero que representa uma idade em anos: "))
    if idade == 0:
        break
    idades.append(idade)

media_idade = round(sum(idades) / len(idades),2)
print(f"A idade média é {media_idade} anos")
```

5) Para um número indeterminado de alunos faça:
- Leia a identificação de cada aluno;
- Calcule a média ponderada, de cada aluno, de suas notas N1 e N2, considerando peso 4 para N1 e peso 6 para N2;
- Determine a Maior média e a Segunda maior média dentre todos os alunos;
- Escreva apenas a identificação e a média dos dois melhores alunos.
Obs.: Considere que nunca haverá empate.

```python
alunos = {}

def calcular_media_ponderada(n1, n2):
    return ((n1 * 4 + n2 * 6) / (4+6))

while True:
    identificacao = input("Digite a identificação do aluno ou ( sair ) para finalizar: ").lower()

    if identificacao =='sair':
        break

    n1 = float(input("Digite a nota N1 do aluno: "))
    n2 = float(input("Digite a nota N2 do aluno: "))

    media_ponderada = calcular_media_ponderada(n1, n2)
    alunos[identificacao] = media_ponderada

if alunos:
    melhores_alunos = [(aluno, nota) for aluno, nota in alunos.items()]

    ranqueamento = sorted(melhores_alunos, key=lambda x: x[1], reverse=True)

    print("Melhores alunos:")

    for identificacao, media in ranqueamento[:2]:
        print(f"{identificacao}: {media}")

else:
    print("Nenhum aluno foi cadastrado.")
```

6) Faça um programa que leia N notas de um aluno e calcule e imprima a média de notas desse aluno, faça isso para uma turma de M alunos.

```python
turma = {}

while True:
    print("\nDeixe em branco para sair")

    aluno = input("Digite o nome do aluno para adicionar: ")

    if aluno == "":
        break

    notas = []

    while True:
      entrada = input("\nDigite a nota do aluno (ou deixe em branco para sair): ")

      if not entrada:
          break

      try:
          nota = float(entrada)
          notas.append(nota)

      except ValueError:
          print("Entrada inválida. Digite um número de ponto flutuante válido.")

    media = (sum(notas) / len(notas) if notas else 0)

    turma[aluno] = media

for aluno, media in turma.items():
    print(f'A média do aluno(a) {aluno} é {media:.2f} pontos')
```

7) Faça um programa que leia N notas de um aluno e calcule e imprima a média de notas desse aluno e sua maior nota, faça isso para uma turma de M alunos. Calcule a maior média entre todos os alunos.

```python
turma = {}

while True:
    print("\nDeixe em branco para sair")

    aluno = input("Digite o nome do aluno para adicionar: ")

    if aluno == "":
        break

    notas = []

    while True:
      entrada = input("\nDigite a nota do aluno (ou deixe em branco para sair): ")

      if not entrada:
          break

      try:
          nota = float(entrada)
          notas.append(nota)

      except ValueError:
          print("Entrada inválida. Digite um número de ponto flutuante válido.")

    media = (sum(notas) / len(notas) if notas else 0)

    maiorNota = max(notas)

    turma[aluno] = {"media": media, "maiorNota": maiorNota}

    maiorMedia = max(turma.values(), key=lambda informacoes: informacoes["media"])["media"]

for aluno, informacoes in turma.items():
    print(f'\n A média do(a) aluno(a) {aluno} é {informacoes["media"]:.2f} pontos e a sua maior nota é {informacoes["maiorNota"]}')

print(f'\n A maior média é de {maiorMedia:.2f} pontos pertecente a {aluno}')
```

8) Faça um programa que leia a altura e o gênero de N pessoas e calcule e imprima a média de altura das mulheres.

```python
generos = []
alturas = []
alturaMulheres = []

while True:
    print("Digite'sair' para finalizar: ")
    genero = input("Digite seu gênero [ Homem ou Mulher ]: ").lower()

    if genero == "sair":
        break

    if genero not in ["homem", "mulher"]:
        print("Entrada inválida. Tente novamente!")
        continue

    altura = int(input("Digite sua Altura em cm: "))

    generos.append(genero)
    alturas.append(altura)

    if genero == "mulher":
        alturaMulheres.append(altura)

if alturaMulheres:
    media = round(sum(alturaMulheres) / len(alturaMulheres), 2)
    print(f'\nMédia da altura das Mulheres é {media} cm')
else:
    print("Nenhuma mulher foi cadastrada.")
```
