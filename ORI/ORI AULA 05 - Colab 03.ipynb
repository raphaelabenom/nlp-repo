1. Faça um programa que receba N strings de entrada lidas do teclado e a imprima convertida em letras minúsculas em formato string. As entrada são lidas, validadas e convertidas uma por vez. Caso uma leitura do teclado possua números, dê uma mensagem: "essa entrada só recebe caracteres não numéricos". O programa pára de funcionar quando a entrada digitada for '***'.

```python
def converterMinuscula():
    while True:
        entrada = input("Digite um texto (ou *** para sair): ")
        if entrada == '***':
            break
        if not entrada.replace(" ", "").isalpha():
            print("essa entrada só recebe caracteres não numéricos")
            continue

        converterTexto = entrada.lower()

        print(converterTexto)

converterMinuscula()
```

2. Faça um programa que leia duas strings A e B e verifique quantas vezes uma palavra de A repete em B (0, 1, 2, ... ou n repetições). Considere palavra uma sequência de caracteres entre espaços brancos,
ou que inicie a string lida seguida de um espaço branco, ou finalize com um espaço branco seguido de palavra. Antes de efetuar tal procedimento, retire do texto os símbolos de pontuação (. , ! ? : ; - /).

```python
import re

def tratamentoTexto(texto: str) -> list:
    texto_sem_simbolos = re.sub(r'[,\!?:;\-/.]', '', texto)
    texto_minusculo = texto_sem_simbolos.lower()
    listaPalavras = texto_minusculo.split()
    return listaPalavras

def contarRepeticoes(textoA: list, textoB: list) -> dict:
    contagem = {}

    for palavraA in textoA:

        contador = 0

        for palavraB in textoB:
            if palavraA == palavraB:
                contador += 1
        if contador > 0:
            contagem[palavraA] = contador

    return contagem

a = input("Digite um texto A para comparação B: ")
b = input("Digite um texto B para comparação A: ")

textoA = tratamentoTexto(a)
textoB = tratamentoTexto(b)

contagem_repeticoes = contarRepeticoes(textoA, textoB)
print("Contagem de repetições: ")
for palavra, repeticoes in contagem_repeticoes.items():
    print(f"A palavra '{palavra}' se repete {repeticoes} vezes em B.")
```

3. Faça um programa que receba duas listas A e B de inteiros, sendo as listas de qualquer tamanho, se de tamanhos diferentes a menor deve ser preenchida com números 1 até igualar em tamanho à maior. O programa deve:

- a) Imprimir a maior soma entre pares de números, os pares são um da lista A e um da lista B, fazendo isso com todos da lista A.

- b) Depois deve verificar nas listas A e B qual deles multiplicados por 3 é maior que a maior soma dos pares identificada anteriormente(se acontecer, senão dê uma mensagem de não ocorrência). Imprimir a lista, o índice e o número inteiro da lista em que ocorreu o fenômeno.

```python
def preencher_listas(lista_a: list, lista_b: list) -> list:
    tamanho_max = max(len(lista_a), len(lista_b))

    for i in range(len(lista_a), tamanho_max):
        lista_a.append(1)

    for i in range(len(lista_b), tamanho_max):
        lista_b.append(1)

    return lista_a, lista_b

def numero_maior_que_multiplicado_por_3(lista_a: list, lista_b: list, maior_soma: int) -> None:
    for i in range(len(lista_a)):
        if lista_a[i] * 3 > maior_soma:
            return lista_a, i, lista_a[i]

    for i in range(len(lista_b)):
        if lista_b[i] * 3 > maior_soma:
            return lista_b, i, lista_b[i]

    return None, None, None

def maior_soma_pares(lista_a: list, lista_b: list) -> int:
    if not lista_a or not lista_b:
        raise ValueError("As listas 'lista_a' e 'lista_b' não podem estar vazias.")

    maior_soma = float('-inf')

    for a in lista_a:
        for b in lista_b:
            if a % 2 == 0 and b % 2 == 0:
                soma = a + b
                if soma > maior_soma:
                    maior_soma = soma

    return maior_soma

lista_a = [1, 2, 3, 4, 5]
lista_b = [6, 7, 8]


lista_a, lista_b = preencher_listas(lista_a, lista_b)
maior_soma = maior_soma_pares(lista_a, lista_b)


lista_encontrada, indice_encontrado, numero_encontrado = numero_maior_que_multiplicado_por_3(lista_a, lista_b, maior_soma)


print(lista_a)
print(lista_b)

print("Maior soma entre pares:", maior_soma)

if lista_encontrada is not None:
    print(f"Número {numero_encontrado} na lista {lista_encontrada} no índice {indice_encontrado} é maior que {maior_soma} quando multiplicado por 3.")
else:
    print("Não há nenhum número que, multiplicado por 3, seja maior que a maior soma dos pares.")
```

4. Faça um programa que receba um texto (nome completo de uma pessoa) considere que ele não tem quebras de linha e somente caracteres alfabéticos. Retire todos os espaços maiores que um entre palavras, e reconstrua esse texto com espaço único entre palavras. O texto de entrada também pode conter n espaços no inicio ou fim do texto que devem ser retirados.

```python
def removerMultiplosEspacos(text: str) -> str:
    palavras = text.split()
    return ' '.join(palavras)

texto = input("Digite um texto: ")
texto = removerMultiplosEspacos(texto)
print(texto)
```

5. Faça um programa que receba três listas com n tuplas em cada lista, e encontre a lista com a maior e menor tupla, imprimindo esse resultado se ele existir.

```python
def encontrarMaiorTupla(list1: list, list2: list, list3: list) -> list:
    lists = [list1, list2, list3]
    max_tuple = None
    min_tuple = None
    max_list = None
    min_list = None

    for lst in lists:
        for tup in lst:
            if max_tuple is None or tup > max_tuple:
                max_tuple = tup
                max_list = lst
            if min_tuple is None or tup < min_tuple:
                min_tuple = tup
                min_list = lst

    if max_list is not None:
        print(f"Lista com a maior tupla: {max_list} - Tupla: {max_tuple}")
    else:
        print("Não há tuplas nas listas")

    if min_list is not None:
        print(f"Lista com a menor tupla: {min_list} - Tupla: {min_tuple}")
    else:
        print("Não há tuplas nas listas")


list1 = [(1, 2), (3, 4), (5, 6)]
list2 = [(7, 8), (9, 10), (11, 12)]
list3 = [(13, 14), (15, 16), (17, 18)]

encontrarMaiorTupla(list1, list2, list3)
```

6. Faça um programa que lê do teclado o nome, peso e a altura de N pessoas. Cada um desses atributos de pessoa (nome, peso e altura) deve ser uma lista, e o índice de cada lista indica o nome, o peso e altura de uma pessoa. O pragrama deve:

- a) Encontrar a pessoa mais alta e imprimir todos os seus dados (nome, peso, altura). Porém, se houver empate na altura mais alta, ele deve imprimir a mais alta e de maior peso. Caso haja ainda haja empate, imprima as N empatadas.

- Obs. Teste seus programas com vários casos de entrada.

```python
def encontrarPessoaMaisAlta(pessoas:int) -> dict:
    if not pessoas:
        raise ValueError("A lista de pessoas não pode estar vazia.")

    pessoa_mais_alta = None
    pessoas_mais_altas = []
    altura_maxima = 0

    for pessoa in pessoas:
        nome, peso, altura = pessoa
        if altura > altura_maxima:
            altura_maxima = altura
            pessoa_mais_alta = pessoa
            pessoas_mais_altas = [pessoa]
        elif altura == altura_maxima:
            if peso > pessoa_mais_alta[1]:
                pessoa_mais_alta = pessoa
                pessoas_mais_altas = [pessoa]
            elif peso == pessoa_mais_alta[1]:
                pessoas_mais_altas.append(pessoa)

    if len(pessoas_mais_altas) == 1:
        print("Pessoa mais alta:")
        print("Nome:", pessoa_mais_alta[0])
        print("Peso:", pessoa_mais_alta[1])
        print("Altura:", pessoa_mais_alta[2])
    else:
        print("Pessoas mais altas e com maior peso:")
        pessoas_mais_altas.sort(key=lambda x: (-x[2], -x[1]))
        for pessoa in pessoas_mais_altas:
            print("Nome:", pessoa[0])
            print("Peso:", pessoa[1])
            print("Altura:", pessoa[2])


n = int(input("Quantas pessoas deseja inserir? "))

pessoas = []

for i in range(n):
    nome = input("Nome da pessoa: ")
    peso = float(input("Peso da pessoa: "))
    altura = float(input("Altura da pessoa: "))
    pessoas.append([nome, peso, altura])


print("\n")
encontrarPessoaMaisAlta(pessoas)
```
